#### urllib库

```python
import urllib.request
url='https://blog.csdn.net/'
reponse=urllib.request.urlopen(url)
```

```python
reponse.status
```

```python
reponse.reason
```

```python
for k, v in reponse.getheaders():
    print('%s: %s' % (k, v))
```

```python
import urllib.request
url='https://blog.csdn.net/'
header={
   'User-Agent':'Mozilla/5.0 (X11; Fedora; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}
request=urllib.request.Request(url,headers=header)
reponse=urllib.request.urlopen(request)
```

```python
import urllib.request
import urllib.parse

url='https://blog.csdn.net/'
header={
   'User-Agent':'Mozilla/5.0 (X11; Fedora; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36'
}

data={'name':'soyoger','pass':'123456'}
postdata=urllib.parse.urlencode(data).encode('utf8')
request=urllib.request.Request(url,postdata)
reponse=urllib.request.urlopen(request).read()

fhandle=open("csdn.html","wb")
fhandle.write(reponse)
fhandle.close()
```

```python
proxy_handler = urllib.request.ProxyHandler({'http': 'http://jjs.ss.uhaiba.top:118'})  #为了安全，自己设置一个，这个地方我把我的能用了改变了，不选择开放
proxy_auth_handler = urllib.request.ProxyBasicAuthHandler()
opener = urllib.request.build_opener(proxy_handler, proxy_auth_handler)
reponse = opener.open('https://blog.csdn.net')
```

```python
reponse.status
```
